```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.svm import SVC
import pandas as pd

# Define parameter grid
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 0.1, 0.01, 0.001],
    'kernel': ['rbf', 'poly', 'sigmoid']
}

# Create the model
svm = SVC()

# Load the data
data = pd.read_csv('Dry_Eye_Dataset - Copy.csv')

# Define target and features
y = data['Dry Eye Disease']
X = data[['Gender', 'Age', 'Sleep duration', 'Sleep quality','Physical activity', 'Daily steps',
       'Sleep disorder', 'Caffeine consumption', 'Alcohol consumption',
       'Smoking','Redness in eye','Discomfort Eye-strain', 'Blue-light filter']]

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Define boolean columns for encoding
bool_cols = [
    'Smoking','Alcohol consumption','Sleep disorder','Caffeine consumption','Redness in eye','Discomfort Eye-strain', 'Blue-light filter'
]

# Encode boolean columns
for col in bool_cols:
    X_train[col] = X_train[col].map({'N': 0, 'Y': 1})
    X_test[col] = X_test[col].map({'N': 0, 'Y': 1})

# Encode gender column
X_train['Gender'] = X_train['Gender'].map({'M': 0, 'F': 1})
X_test['Gender'] = X_test['Gender'].map({'M': 0, 'F': 1})

# Encode target variable
y_train = y_train.map({'N': 0, 'Y': 1})
y_test = y_test.map({'N': 0, 'Y': 1})

# Grid search with 5-fold cross validation
grid_search = GridSearchCV(estimator=svm, param_grid=param_grid, cv=5, scoring='accuracy', verbose=1)
grid_search.fit(X_train, y_train)

# Best parameters and accuracy
print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_ * 100)

# Use best model for predictions
best_svm = grid_search.best_estimator_
y_pred_svm = best_svm.predict(X_test)

print("\nFinal Test Accuracy:", accuracy_score(y_test, y_pred_svm) * 100)
```
